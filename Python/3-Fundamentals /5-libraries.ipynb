## 1️⃣ Random Choice

```python
import random
print(random.choice(["heads", "tails"]))
```

## 2️⃣ randint

```python
import random
print(random.randint(1, 10))
```

## 3️⃣ shuffle

```python
import random

cards = ["jack", "queen", "king"]
random.shuffle(cards)

for card in cards:
    print(card)
```

## 4️⃣ Statistics

```python
import statistics
print(statistics.mean([100, 90]))
```

## 5️⃣ sys.argv (CLI Input) ⭐

```python
import sys

if len(sys.argv) != 2:
    sys.exit("Usage: python file.py name")

print("hello,", sys.argv[1])
```

## 6️⃣ Loop over CLI args

```python
import sys

for arg in sys.argv[1:]:
    print("hello,", arg)
```

## 7️⃣ API Request (requests) ⭐

```python
import requests

response = requests.get("https://api.github.com")
print(response.status_code)
```

## 8️⃣ JSON Formatting

```python
import requests, json

response = requests.get("https://api.github.com")
print(json.dumps(response.json(), indent=2))
```

## 9️⃣ Loop Through JSON

```python
import requests

response = requests.get("https://itunes.apple.com/search?entity=song&term=adele")
data = response.json()

for item in data["results"]:
    print(item["trackName"])
```
