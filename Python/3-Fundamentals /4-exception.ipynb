- ValueError / NameError

- loops + validation

## 1️⃣ Basic Input

```python
x = int(input("What's x? "))
print(f"x is {x}")
```

## 2️⃣ Handle ValueError

```python
try:
    x = int(input("What's x? "))
    print(f"x is {x}")
except ValueError:
    print("Invalid input")
```

## 3️⃣ Using else

```python
try:
    x = int(input("What's x? "))
except ValueError:
    print("Invalid input")
else:
    print(f"x is {x}")
```

## 4️⃣ Loop Until Valid Input ⭐

```python
while True:
    try:
        x = int(input("What's x? "))
        break
    except ValueError:
        print("Invalid input")

print(f"x is {x}")
```

## 5️⃣ Function Version ⭐

```python
def get_int():
    while True:
        try:
            return int(input("What's x? "))
        except ValueError:
            print("Invalid input")

x = get_int()
print(f"x is {x}")
```

## 6️⃣ Clean Version (No print in except)

```python
def get_int():
    while True:
        try:
            return int(input("What's x? "))
        except ValueError:
            pass

x = get_int()
print(f"x is {x}")
```

## 7️⃣ Using pass

```python
def get_int():
    while True:
        try:
            return int(input("What's x? "))
        except ValueError:
            pass  # تجاهل الخطأ وكرر

print(get_int())
```

## 8 Reusable Function with Prompt ⭐

```python
def get_int(prompt):
    while True:
        try:
            return int(input(prompt))
        except ValueError:
            pass

x = get_int("Enter a number: ")
print(f"x is {x}")
```
