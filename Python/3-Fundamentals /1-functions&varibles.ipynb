## 1️⃣ Hello World

```python
print("hello, world")
```

## 2️⃣ Input + Print

```python
name = input("What's your name? ")
print("hello,")
print(name)
```

## 3️⃣ String Concatenation

```python
name = input("What's your name? ")
print("hello, " + name)
```

## 4️⃣ Multiple Arguments

```python
name = input("What's your name? ")
print("hello,", name)
```

## 5️⃣ end Parameter

```python
name = input("What's your name? ")
print("hello, ", end="")
print(name)
```

## 6️⃣ f-Strings ⭐

```python
name = input("What's your name? ")
print(f"hello, {name}")
```

## 7️⃣ String Cleaning

```python
name = input("What's your name? ").strip().title()
print(f"hello, {name}")
```

## 8️⃣ Split Name

```python
name = input("What's your name? ").strip().title()
first, last = name.split(" ")
print(f"hello, {first}")
```

## 9️⃣ Basic Math

```python
x = 1
y = 2
print(x + y)
```

## 🔟 Common Mistake (String Concatenation ❌)

```python
x = input("What's x? ")
y = input("What's y? ")
print(x + y)
```

## 1️⃣1️⃣ Convert to int ✅

```python
x = int(input("What's x? "))
y = int(input("What's y? "))
print(x + y)
```

## 1️⃣2️⃣ Float Numbers

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
print(x + y)
```

## 1️⃣3️⃣ Rounding

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
print(round(x + y))
```

## 1️⃣4️⃣ Formatting Numbers

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
z = round(x + y)
print(f"{z:,}")
```

## 1️⃣5️⃣ Division

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
print(x / y)
```

## 1️⃣6️⃣ Format Decimals

```python
x = int(input("What's x? "))
y = int(input("What's y? "))
print(f"{x / y:.2f}")
```

## 1️⃣7️⃣ Functions

```python
def hello():
    print("hello")

hello()
```

## 1️⃣8️⃣ Function with Parameter

```python
def hello(to):
    print("hello,", to)

hello("Rasha")
```

## 1️⃣9️⃣ Default Parameter

```python
def hello(to="world"):
    print("hello,", to)

hello()
hello("Rasha")
```

## 2️⃣0️⃣ main() Pattern ⭐

```python
def main():
    name = input("What's your name? ")
    hello(name)

def hello(to="world"):
    print("hello,", to)

main()
```

## 2️⃣1️⃣ Return Value ⭐

```python
def square(n):
    return n * n

x = int(input("What's x? "))
print("x squared is", square(x))
```
