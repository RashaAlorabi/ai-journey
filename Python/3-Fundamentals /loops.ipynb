# 🧠 Python Loops + Data Structures + Patterns

- while / for loops
- break / continue
- lists & dictionaries
- nested loops (مهم جدًا 🔥)
- patterns (أسئلة مقابلات كلاسيكية)

## 1️⃣ Repetition بدون Loop (❌ غير عملي)

```python
print("meow")
print("meow")
print("meow")
```

## 2️⃣ while Loop

```python
i = 0
while i < 3:
    print("meow")
    i += 1
```

## 3️⃣ for Loop (أفضل غالبًا) ⭐

```python
for _ in range(3):
    print("meow")
```

## 4️⃣ String Multiplication

```python
print("meow\n" * 3, end="")
```

## 5️⃣ break / continue

```python
while True:
    n = int(input("Enter n (>0): "))
    if n > 0:
        break

for _ in range(n):
    print("meow")
```

## 6️⃣ Functions + Loop

```python
def meow(n):
    for _ in range(n):
        print("meow")

meow(3)
```

## 7️⃣ Lists

```python
students = ["Hermione", "Harry", "Ron"]

for student in students:
    print(student)
```

## 8️⃣ Index + List

```python
students = ["Hermione", "Harry", "Ron"]

for i in range(len(students)):
    print(i + 1, students[i])
```

## 9️⃣ Dictionaries

```python
students = {
    "Hermione": "Gryffindor",
    "Harry": "Gryffindor",
    "Ron": "Gryffindor",
    "Draco": "Slytherin"
}

for student in students:
    print(student, students[student], sep=", ")
```

## 🔟 List of Dictionaries (🔥 مهم جدًا)

```python
students = [
    {"name": "Hermione", "house": "Gryffindor"},
    {"name": "Harry", "house": "Gryffindor"},
    {"name": "Ron", "house": "Gryffindor"},
]

for student in students:
    print(student["name"], student["house"])
```

## 1️⃣1️⃣ Patterns (Column)

```python
for _ in range(3):
    print("#")
```

## 1️⃣2️⃣ Patterns (Row)

```python
print("#" * 5)
```

## 1️⃣3️⃣ Patterns (Square - Nested Loop) 🔥

```python
size = 3

for i in range(size):
    for j in range(size):
        print("#", end="")
    print()
```

## 1️⃣4️⃣ Patterns (Optimized)

```python
size = 3

for _ in range(size):
    print("#" * size)
```
