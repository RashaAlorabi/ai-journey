## 1️⃣ Function to Test

```python
def square(n):
    return n * n
```

## 2️⃣ Basic Testing (Manual ❌)

```python
def test_square():
    if square(2) != 4:
        print("Error")
```

## 3️⃣ Using assert ✅

```python
def test_square():
    assert square(2) == 4
    assert square(3) == 9
```

## 4️⃣ Multiple Test Cases ⭐

```python
def test_positive():
    assert square(1) == 1
    assert square(2) == 4

def test_negative():
    assert square(-2) == 4

def test_zero():
    assert square(0) == 0
```

## 5️⃣ Testing Errors (pytest.raises) ⭐

```python
import pytest

def test_type_error():
    with pytest.raises(TypeError):
        square("cat")
```

## 6️⃣ Example: hello() Function

```python
def hello(to="world"):
    return f"hello, {to}"
```

## 7️⃣ Testing hello() ⭐

```python
def test_default():
    assert hello() == "hello, world"

def test_argument():
    assert hello("Rasha") == "hello, Rasha"
```

## 8️⃣ Multiple Inputs

```python
def test_multiple_names():
    for name in ["Ali", "Sara", "Omar"]:
        assert hello(name) == f"hello, {name}"
```
