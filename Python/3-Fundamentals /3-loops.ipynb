## 1️⃣ Repetition (Wrong Way ❌)

```python
print("meow")
print("meow")
print("meow")
```

## 2️⃣ while Loop

```python
i = 0
while i < 3:
    print("meow")
    i += 1
```

## 3️⃣ for Loop ⭐

```python
for _ in range(3):
    print("meow")
```

## 4️⃣ String Multiplication

```python
print("meow\n" * 3, end="")
```

## 5️⃣ break / continue

```python
while True:
    n = int(input("Enter n (>0): "))
    if n > 0:
        break

for _ in range(n):
    print("meow")
```

## 6️⃣ Functions + Loop

```python
def meow(n):
    for _ in range(n):
        print("meow")

meow(3)
```

## 7️⃣ Lists

```python
students = ["Hermione", "Harry", "Ron"]

for student in students:
    print(student)
```

## 8️⃣ Index + List

```python
students = ["Hermione", "Harry", "Ron"]

for i in range(len(students)):
    print(i + 1, students[i])
```

## 9️⃣ Dictionary

```python
students = {
    "Hermione": "Gryffindor",
    "Harry": "Gryffindor",
    "Ron": "Gryffindor",
    "Draco": "Slytherin"
}

for name in students:
    print(name, students[name])
```

## 🔟 List of Dictionaries ⭐

```python
students = [
    {"name": "Hermione", "house": "Gryffindor"},
    {"name": "Harry", "house": "Gryffindor"},
    {"name": "Ron", "house": "Gryffindor"},
]

for student in students:
    print(student["name"], student["house"])
```

## 1️⃣1️⃣ Pattern: Column

```python
for _ in range(3):
    print("#")
```

## 1️⃣2️⃣ Pattern: Row

```python
print("#" * 5)
```

## 1️⃣3️⃣ Pattern: Square (Nested Loop) 🔥

```python
size = 3

for i in range(size):
    for j in range(size):
        print("#", end="")
    print()
```

## 1️⃣4️⃣ Optimized Pattern

```python
size = 3

for _ in range(size):
    print("#" * size)
```

```python
# Introduces continue, break

while True:
    try:
        n = int(input("What's n? "))
        if n <= 0:
            continue
        else:
            break
    except ValueError:
        print("Please enter a valid number")

for _ in range(n):
    print("meow")
```

```python
# Removes continue

while True:
    try:
        n = int(input("What's n? "))
        if n > 0:
            break
    except ValueError:
        print("Please enter a valid number")

for _ in range(n):
    print("meow")
```
