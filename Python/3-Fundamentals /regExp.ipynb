# Email Validation & Regex Practice

```python
# validate0.py
email = input("What's your email? ").strip()

if "@" in email:
    print("Valid")
else:
    print("Invalid")
```

```python
# validate1.py
email = input("What's your email? ").strip()

if "@" in email and "." in email:
    print("Valid")
else:
    print("Invalid")
```

```python
# validate2.py
email = input("What's your email? ").strip()

username, domain = email.split("@")

if username and "." in domain:
    print("Valid")
else:
    print("Invalid")
```

```python
# validate3.py
email = input("What's your email? ").strip()

username, domain = email.split("@")

if username and domain.endswith(".edu"):
    print("Valid")
else:
    print("Invalid")
```

```python
# validate11.py
import re

email = input("What's your email? ").strip()

if re.search(r"^\w+@\w+\.edu$", email, re.IGNORECASE):
    print("Valid")
else:
    print("Invalid")
```

# Name Formatting

```python
# format3.py
import re

name = input("What's your name? ").strip()
if matches := re.search(r"^(.+), (.+)$", name):
    name = matches.group(2) + " " + matches.group(1)

print(f"hello, {name}")
```

# Twitter Username Extraction

```python
# Extracts Twitter username from URL using str.replace

url = input("URL: ").strip()
username = url.replace("https://twitter.com/", "")
print(f"Username: {username}")
```

```python
# Extracts Twitter username from URL using str.removeprefix

url = input("URL: ").strip()
username = url.removeprefix("https://twitter.com/")
print(f"Username: {username}")
```

```python

import re

url = input("URL: ").strip()

username = re.sub(r"^(https?://)?(www\.)?twitter\.com/", "", url)

print(f"username:{username}")
```

```python
# twitter5.py
import re

url = input("URL: ").strip()

matches = re.search(r"^https?://(?:www\.)?twitter\.com/(.+)$", url, re.IGNORECASE)

if matches:
    print(f"Username: {username}")
```

```python
import re
url = input("URL: ").strip()
matches = re.search(r"^https?://(?:www\.)?twitter\.com/([a-z0-9_]+)", url, re.IGNORECASE)
if matches:
    print("Username:", matches.group(1))
```
