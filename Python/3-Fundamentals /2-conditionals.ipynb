## 1️⃣ Basic Conditions

```python
x = int(input("x: "))
y = int(input("y: "))

if x < y:
    print("x < y")
if x > y:
    print("x > y")
if x == y:
    print("x == y")
```

## 2️⃣ if / elif / else

```python
if x < y:
    print("x < y")
elif x > y:
    print("x > y")
else:
    print("x == y")
```

## 3️⃣ Logical Operators

```python
if x != y:
    print("Not equal")
else:
    print("Equal")
```

## 4️⃣ Grading System ⭐

```python
score = int(input("Score: "))

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

## 5️⃣ String Comparison

```python
answer = input("Do you agree? ")

if answer == "yes":
    print("Agreed")
else:
    print("Not agreed")
```

## 6️⃣ Best Practice (strip + lower) ⭐

```python
answer = input("Do you agree? ").strip().lower()

if answer.startswith("y"):
    print("Agreed")
else:
    print("Not agreed")
```

## 7️⃣ Even / Odd

```python
x = int(input("Enter number: "))

if x % 2 == 0:
    print("Even")
else:
    print("Odd")
```

## 8️⃣ Function Returning Boolean ⭐

```python
def is_even(n):
    return n % 2 == 0

x = int(input("Enter number: "))

if is_even(x):
    print("Even")
else:
    print("Odd")
```

## 9️⃣ match-case (Python 3.10+)

```python
name = input("Name: ")

match name:
    case "Harry" | "Hermione" | "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Unknown")
```
