# 🧠 Python Conditionals (if / elif / else)



## 1️⃣ Basic if Conditions
- تنفيذ شرط واحد أو أكثر
- كل if يتم تقييمه بشكل مستقل

```python
x = int(input("Enter x: "))
y = int(input("Enter y: "))

if x < y:
    print("x is less than y")

if x > y:
    print("x is greater than y")

if x == y:
    print("x is equal to y")
```

## 2️⃣ if / elif / else (Mutually Exclusive)
- يتم تنفيذ شرط واحد فقط

```python
if x < y:
    print("x < y")
elif x > y:
    print("x > y")
else:
    print("x == y")
```

## 3️⃣ Logical Operators
- and / or / not

```python
if x < y or x > y:
    print("Not equal")
else:
    print("Equal")
```

## 4️⃣ Comparison Operators
- == (equal)
- != (not equal)

```python
if x != y:
    print("Not equal")
else:
    print("Equal")
```

## 5️⃣ Grading System (Important for Interviews)
- مثال كلاسيكي جدًا

```python
score = int(input("Score: "))

if score >= 90:
    print("A")
elif score >= 80:
    print("B")
elif score >= 70:
    print("C")
elif score >= 60:
    print("D")
else:
    print("F")
```

## 6️⃣ String Handling (Important ⚠️)
- strip() لإزالة الفراغات
- lower() لتوحيد الحروف

```python
answer = input("Do you agree? ").strip().lower()

if answer.startswith("y"):
    print("Agreed")
else:
    print("Not agreed")
```

## 7️⃣ Even / Odd (Functions + Clean Code)
- مهم جدًا في المقابلات

```python
def is_even(n):
    return n % 2 == 0

x = int(input("Enter number: "))

if is_even(x):
    print("Even")
else:
    print("Odd")
```

## 8️⃣ match-case (Python 3.10+)
- بديل switch

```python
name = input("Name: ")

match name:
    case "Harry" | "Hermione" | "Ron":
        print("Gryffindor")
    case "Draco":
        print("Slytherin")
    case _:
        print("Unknown")
```
