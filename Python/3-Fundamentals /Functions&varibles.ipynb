## 1️⃣ Basic Printing

```python
print("hello, world")
```

## 2️⃣ Input + String Output

```python
name = input("What's your name? ")
print("hello,", name)
```

## 3️⃣ String Concatenation

```python
name = input("What's your name? ")
print("hello, " + name)
```

## 4️⃣ f-String (أفضل طريقة) ⭐

```python
name = input("What's your name? ")
print(f"hello, {name}")
```

## 5️⃣ String Cleaning

```python
name = input("What's your name? ").strip().title()
print(f"hello, {name}")
```

## 6️⃣ Splitting Strings

```python
name = input("What's your name? ").strip().title()
first, last = name.split(" ")
print(f"hello, {first}")
```

## 7️⃣ Basic Math

```python
x = 1
y = 2
z = x + y
print(z)
```

## 8️⃣ ⚠️ Common Mistake (Strings vs Numbers)

```python
x = input("What's x? ")
y = input("What's y? ")
print(x + y)  # Concatenation ❌
```

## 9️⃣ Fix: Convert to int

```python
x = int(input("What's x? "))
y = int(input("What's y? "))
print(x + y)
```

## 🔟 Float Numbers

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
print(x + y)
```

## 1️⃣1️⃣ Rounding

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
print(round(x + y))
```

## 1️⃣2️⃣ Formatting Numbers

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
z = round(x + y)
print(f"{z:,}")
```

## 1️⃣3️⃣ Division

```python
x = float(input("What's x? "))
y = float(input("What's y? "))
print(x / y)
```

## 1️⃣4️⃣ Formatting Decimals

```python
x = int(input("What's x? "))
y = int(input("What's y? "))
z = x / y
print(f"{z:.2f}")
```

## 1️⃣5️⃣ Functions Basics

```python
def hello():
    print("hello")

hello()
```

## 1️⃣6️⃣ Function with Parameter

```python
def hello(to):
    print("hello,", to)

name = input("What's your name? ")
hello(name)
```

## 1️⃣7️⃣ Default Parameter

```python
def hello(to="world"):
    print("hello,", to)

hello()
hello("Rasha")
```

## 1️⃣8️⃣ main() Function Pattern ⭐

```python
def main():
    name = input("What's your name? ")
    hello(name)

def hello(to="world"):
    print("hello,", to)

main()
```

## 1️⃣9️⃣ Function with Return

```python
def square(n):
    return n * n

x = int(input("What's x? "))
print("x squared is", square(x))
```
